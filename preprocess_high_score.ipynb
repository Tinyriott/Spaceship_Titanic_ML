# 데이터 전처리

## 파생변수 만들기

```python
df
```

### GroupSize 와 IsAlone 구하기

```python
tmp_df = df.copy()
```

```python
tmp_df
```

```python
new_passenger_id = []
for c in df["PassengerId"]:
    tmp = ""
    for d in c:
        if d != "_":
            tmp += d
        else:
            new_passenger_id.append(tmp)
            break
```

```python
tmp_df["PassengerId"] = new_passenger_id
```

```python
tmp_df
```

```python
cnt_p_id = Counter(new_passenger_id)
```

```python
group_num = []
for c in tmp_df["PassengerId"]:
    group_num.append(cnt_p_id[c])        
```

```python
len(group_num)
```

```python
tmp_df["group_num"] = group_num
```

```python
tmp_df
```

```python
tmp_df["IsAlone"] = 1
tmp_df.loc[tmp_df["group_num"] > 1, "IsAlone"] = 0
```

```python
tmp_df
```

```python
df["GroupSize"] = tmp_df["group_num"]
df["IsAlone"] = tmp_df["IsAlone"]
```

```python
df
```

### 자리 나누기

```python
deck = []
number = []
side = []

for c in df["Cabin"]:
    if type(c) == float:
        deck.append(np.nan)
        number.append(np.nan)
        side.append(np.nan)
    else:
        a = c.split("/")
        deck.append(a[0])
        number.append(a[1])
        side.append(a[2])
```

```python
df["Cabin_Deck"] = deck
df["Cabin_Num"] = number
df["Cabin_Side"] = side
```

```python
df = df.drop(["Cabin"], axis = 1)
```

## 결측값 처리하기

### 데이터 확인

```python
df
```

```python
df.isnull().sum()
```

### 최빈값으로 채우기

```python
df[["Destination", "VIP", "HomePlanet", "CryoSleep"]] = df[["Destination", "VIP", "HomePlanet", "CryoSleep"]].fillna(df[["Destination", "VIP", "HomePlanet", "CryoSleep"]].mode().iloc[0])
```

```python
df.isnull().sum()
```

```python
df[["RoomService", "FoodCourt", "ShoppingMall", "Spa", "VRDeck"]] = df[["RoomService", "FoodCourt", "ShoppingMall", "Spa", "VRDeck"]].fillna(0)
```

```python
df.isnull().sum()
```

### Cabin 채우기

```python
df
```

```python
for col in ['Cabin_Deck', 'Cabin_Num', 'Cabin_Side']:
    df[col] = df[col].where(df['IsAlone'] == 1, df[col].ffill())
```

```python
df.isnull().sum()
```

```python
df[['Cabin_Deck', 'Cabin_Num', 'Cabin_Side']] = df[['Cabin_Deck', 'Cabin_Num', 'Cabin_Side']].fillna(df[['Cabin_Deck', 'Cabin_Num', 'Cabin_Side']].mode().iloc[0])
```

```python
df.isnull().sum()
```

```python
df["Total_Amenities"] = df["RoomService"] + df["FoodCourt"] + df["ShoppingMall"] + df["Spa"] + df["VRDeck"]
```

```python
df["NoSpending"] = 1
df.loc[df["Total_Amenities"] > 1, "NoSpending"] = 0
```

```python
df
```

### Age 출발지역, 도착지 별 나이 평균으로

```python
df.groupby(["HomePlanet", "Destination"])["Age"].mean()
```

```python
df["Age"] = df["Age"].fillna(df.groupby(["HomePlanet", "Destination"])["Age"].transform("mean"))
```
